## 1. 선(line)으로 시각화하기

### 예제 1 : 1차원 데이터 생성 및 시각화

```python
# Python package 가져오기 및 matplotlib 출력 옵션 설정
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# 주피터 노트북을 실행한 브라우저에서 바로 그림을 볼 수 있게끔 만드는 것. 즉, 브라우저 내부(inline)에 바로 그려지도록 해주는 코드.
```

#### 코드 설명

```python
# 임의의(랜덤) 값 40개로 이루어진 배열 변수 생성
value = np.random.standard_normal(40)
value
```

#### 변수 설명
- value (y축 데이터) : 40개의 랜덤 값으로 이루어진 *numpy.ndarray형 변수

#### 패키지 설명
- numpy.random : numpy의 랜덤 값 생성 관련 함수를 모아놓은 패키지

- numpy.random.standard_normal() : 표준 정규 분포를 따르는 랜덤 값을 생성하는 함수

```python
# 선 plot 그리기 및 x, y축 범위 설정
index = range(len(value))
plt.plot(index, value)
# x.y 축 범위 설정
plt.xlim(0, 20)
plt.ylim(np.min(value)-1, np.max(value)+1)
```

#### 변수 설명

### 예제 2 : 2차원 데이터 생성 및 시각화

```python
# 랜덤 값으로 이루어진 2차원 배열 변수 생성하기
value = np.random.standard_normal((30, 2))
value
```

#### 변수 설명
- value (y축 데이터) : 30 x 2개의 랜덤 값들로 이루어진 정규표준분포를 따르는 numpy.ndarray형 변수

```python
# 한 그래프에 그리기
plt.figure(figsize=(10, 4))
plt.plot(value[:, 0], lw=1.5, label='1st')
plt.plot(value[:, 1], lw=1.5, label='2nd')
plt.plot(value, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 2')
```

#### 함수 설명
- pyplot.grid() : grid(격자)로 표시 유무 설정
- pyplot.legend() : 범례 표시 함수

```python
# 데이터 별로 각각 그리기
# 1번째
plt.figure(figsize=(10, 5))
plt.subplot(211)
plt.plot(value[:, 0], lw=1.5, label='1st')
plt.plot(value[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Line Plot 3')
# 2번째
plt.subplot(212)
plt.plot(value[:, 1], lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight') # plt.axis("tight") : 모든 자료가 보이도록 범위를 타이트하게 조정, square(정사각형), scaled(스케일이 같도록) 등 존재
plt.ylabel('value')
```

#### 함수 설명
- pyplot.subplot(nrows ncols plot_number) : 서브플롯을 반환하는 함수
- 파라미터 설명 
1. nrows : 전체 서브플롯 행 개수 
2. ncols : 전체 서브플롯 열 개수 
3. plot_number : 특정 플롯의 번호
- ex) subplot(211) : 2x1개의 서브플롯으로 이루어진 플롯의 첫 번째 서브플롯

```python
# 6개의 각각 다른 서브플롯 그려보기
# 1번째
plt.figure(figsize=(13, 5))
plt.subplot(231)
plt.plot(value[:, 0], lw=1.5, label='1st')
plt.plot(value[:, 0], 'co')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Line Plot 3')
# 2번째
plt.subplot(232)
plt.plot(value[:, 0],'g-.', lw=1.5, label='1st')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
# 3번째
plt.subplot(233)
plt.plot(value[:, 0],'g', lw=1.5, label='1st')
plt.plot(value[:, 0], 'bD')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
# 4번째
plt.subplot(234)
plt.plot(value[:, 1],'*', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
# 5번째
plt.subplot(235)
plt.plot(value[:, 1],'b', lw=1.5, label='2nd')
plt.plot(value[:, 1], 'ms')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
# 6번째
plt.subplot(236)
plt.plot(value[:, 1],'r--', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
```

## 2. 산점도(scatter plot)으로 시각화하기

### 예제 1 : pyplot.plot() 함수를 이용한 산점도 그리기

```python
# 랜덤 값들로 이루어진 2차원 배열 변수 생성 후 Plot 그리기
value = np.random.standard_normal((500, 2))
plt.plot(value[:, 0], value[:, 1], 'ro')
plt.grid(False)
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 1')
```

## 1. 히스토그램(histogram)으로 시각화하기

### 예제 1 : pyplot.hist() 함수를 이용한 히스토그램 그리기

```python
# 히스토그램 그리기
plt.figure(figsize=(12, 7))
plt.hist(value, label=['1st', '2nd'], bins=30)
plt.grid(True)
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

#### 함수 설명
- pyplot.hist() : 히스토그램을 그리는 함수

```python
# 히스토그램 색 입히기 – 데이터 셋 별로 다른 색 입히기
plt.figure(figsize=(12, 6))
plt.hist(value, label=['1st', '2nd'], bins=30, color=['yellow', 'green'])
plt.grid(True)
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 2')
```

```python
# 히스토그램 색 입히기 – 각 patch에 다른 색 입히기
value = np.random.standard_normal(500)
cm = plt.cm.get_cmap('Spectral')
plt.figure(figsize=(12, 6))

# Plot histogram
n, bins, patches = plt.hist(value, bins=30, color='green')
print("1. bins : " + str(bins))
print("2. the length of bins : " + str(len(bins)))
bin_centers = 0.5 * (bins[:-1] + bins[1:])
print("3. bin_centers : " + str(bin_centers))

# scale values to interval [0,1] : 정규화(normalization)
col = (bin_centers - min(bin_centers) / (max(bin_centers) - min(bin_centers)))
print("4. col : " + str(col))

for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))
    
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 3')
plt.show()
```

#### 패키지 설명
- pyplot.cm: 다양한 colormap들과 관련 함수를 제공하는 패키지

#### 함수 설명
- pyplot.cm.get_cmap() : colormap 인스턴스를 가져오는 함수
- zip() : 다른 변수들을 묶어 이를 반복할 수 있는 반복자를 생성하는 함수
- pyplot.setp() : artist object의 옵션을 설정하는 함수 *matplotlib.artist.Artist : 캔버스 관련 object

#### 코드 설명(13번째 줄)
- bin_centers 변수의 값을 0에서 1사이의 값으로 값의 범위를 줄이는 코드

#### 정규화(Normalization)
- 데이터의 전체 구간을 정해진 범위 안으로 변환하여 데이터를 관찰하는 방법

#### 데이터를 정규화하는 방법
- 최소-최대 정규화 : 0(최소값)과 1(최대값) 사이로 데이터의 범위를 변환
- Z-점수 정규화 : 평균 및 표준편차를 기반으로 데이터 조정. 데이터와 평균의 차이를 표준편차로 나눈다.
- 소수점 배열 : 특성 값의 소수점을 이동하여 데이터 크기 조정

#### 최소-최대 정규화 수식
- 새로운 값 = (현재값 - 최소값) / (최대값 - 최소값)

## 2. Seaborn(Statistical Data Visualization)으로 시각화하기

### 예제 1 : "tips" 데이터로 boxplot 그리기

```python
# seaborn 패키지 가져오기, seaborn 스타일 옵션 설정. seaborn 내장 데이터셋 "tips" 로드
import seaborn as sns
# set aesthetic parameters in one step
sns.set(style='whitegrid', palette='pastel')

# Load the example tips dataset
data = sns.load_dataset('tips')
data
```

#### 변수 설명
- data : seaborn 라이브러리의 내장 데이터셋인 “tips”로 pandas.DataFrame형의 변수
- pandas.DataFrame : 2차원의 칼럼으로 라벨링된 데이터 구조
    
#### 함수 설명
- seaborn.set() : 시각화의 미적인 옵션을 한 번에 설정하는 함수
- seaborn.load_dataset() : seasborn 내장 데이터셋을 불러오는 함수

```python
# 요일(day)별 사례금(tip) 박스플롯 그리기
sns.boxplot(x='day', y='tip', hue='sex', data=data, palette='muted')
```

#### 함수 설명
- seaborn.boxplot() : 각 카테고리에 대하여 분포를 보여주기 위한 박스플롯을 그리는 함수

### 예제 2 : "flihgts" 데이터로 heatmap 그리기

```python
# seaborn 내장 데이터셋 "flights" 로드
data2 = sns.load_dataset('flights')
data2
```

```python
# 데이터 pivot 시키기. index : 'year', column : 'month', value : 'passengers'
pivoted_data = data2.pivot("year", 'month', 'passengers')
pivoted_data
```

#### 변수 설명
- pivoted_data : 기존의 data2 변수를 pivot시켜서 재형성한 새로운 pandas.DataFrame형 변수
#### 함수 설명
- pandas.DataFrame.pivot(): pivot 테이블을 생성하여 데이터를 재형성하는 함수

```python
# 스타일 옵션 변경하기 및 Heatmap으로 나타내기
sns.set(context='poster', font='monospace')

sns.heatmap(pivoted_data, annot=True, fmt='d', linewidth=3)
```

#### 함수 설명
- seaborn.heatmap() : 직사각형 형태의 데이터를 색을 입힌 매트릭스 형태의 플롯으로 그리는 함수

```python
sns.set()
sns.reset_orig() # reset params to defaults
sns.heatmap(pivoted_data, annot=True, fmt='d', linewidth=3)
```

#### 함수 설명
- seaborn.reset_orig() : rc 파라미터를 초기 셋팅으로 초기화시키는 함수
- rc 파라미터 : 폰트 옵션 관련 인자
