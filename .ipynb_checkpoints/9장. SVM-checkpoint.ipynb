```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
print(iris.keys())
print(iris.data.shape)
print(iris.feature_names)
```

```python
print(iris.DESCR)
```

```python
x = iris.data[:, :2]
```

```python
x
```

```python
y = iris.target
```

```python
y
```

# 1. Linear 함수를 kernel을 사용하고 C값을 1로 하여 모델 구축(훈련) - 정확도 : 0.82

```python
SVM = svm.SVC(kernel='linear', C=1).fit(x, y)
# 위의 x, y값을 training set으로 하여 SVM모델을 훈련시킴
# kernel로 linear함수를 사용
# C 값을 작게하면 한쪽의 오분류를 낮추는 대신에 다른쪽 오분류가 높아지고, C값을 크게하면 양쪽의 오분류를 비슷하게 허용, 기본값은 1
# fit() : 모델 훈련을 시작하는 함수
```

```python
x_min, x_max = x[:, 0].min()-1, x[:, 0].max()+1
y_min, y_max = x[:, 1].min()-1, x[:, 1].max()+1
plot_unit = 0.025
# x_min과 x_max 사이에 plot_unit(0.025)만큼의 간격을 둔 숫자를 생성
# y_min과 y_max 사이에 plot_unit(0.025)만큼의 간격을 둔 숫자를 생성

#meshgrid() : 첫번째 항을 이용하여 xx에 할당되는 array를 만들어냄. 두번째 항을 이용하여 yy에 할당되는 array를 만들어냄
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_unit), np.arange(y_min, y_max, plot_unit))
```

```python
print("x_min:", x_min)
print("x_max:", x_max)
print("y_min:", y_min)
print("y_max:", y_max)
print("xx:", xx)
print("yy:", yy)
```

```python
z = SVM.predict(np.c_[xx.ravel(), yy.ravel()])
# ravel() 함수는 1차원 배열을 만듦
# np.c_[ , ] : 두 개의 1차원 배열의 대응되는 값으로 구성되는 2차원 배열을 만듦

print(np.c_[xx.ravel(), yy.ravel()])
print()
print("z:", z)
print(z.shape)


z = z.reshape(xx.shape)  # z 값을 xx.shape형태(176 x 224 개)로 변경
print(z.shape)
print(xx.shape)
print(z)
```

```python
plt.pcolormesh(xx, yy, z, alpha=0.1) # xx, yy 좌표에 해당되는 z값에 따라 색상을 다르게 적용. 투명도는 0.1
plt.scatter(x[:, 0],x[:, 1],c=y) # sepal_length와 sepal_width의 산점도 출력
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('Support Vector Machine')
plt.show()
print('정확도 : ',SVM.score(X = x, y = y))
```

# 2. RBF 함수 Kernel을 사용하고 C값을 1로 하여 모델 구축(훈련) - 정확도 : 0.82

```python
SVM = svm.SVC(kernel='rbf', C=1).fit(x, y)
z = SVM.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.pcolormesh(xx, yy, z,alpha=0.1)
plt.scatter(x[:, 0],x[:, 1],c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('Support Vector Machine')
plt.show()
print('정확도 : ',SVM.score(X = x, y = y))
```
